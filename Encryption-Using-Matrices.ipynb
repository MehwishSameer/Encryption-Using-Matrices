```python
import numpy as np

message = str(input("Enter a message of three letter only: "))
```

```python
ENCRYPT_ARR = np.array([[1, 0, 1], [2, 1, 3], [1, 2, 0]])
length = len(message)
to_array = [char for char in message]
print("Message converted to Matrix form: ",to_array)
```

```python
for n in range(length):
    to_array[n] = ord(to_array[n])
    if to_array[n] > 64 & to_array[n] < 91:
        to_array[n] = int((to_array[n]) % 65)
    elif to_array[n] > 96 & to_array[n] < 123:
        to_array[n] = int((to_array[n]) % 97)
print("Matrix converted to numbers using keys as 'A-Z' --> '0-25':", to_array)
```

```python
to_array = np.dot(to_array, ENCRYPT_ARR)
print("Matrix after Multiplication: ", to_array)
```

```python
encrypt = np.empty(length, dtype=object)
for n in range(length):
    if to_array[n]>25:
        to_array[n] = to_array[n] -26
    to_array[n] = to_array[n] + 65
    encrypt[n] = chr(to_array[n])
print("Matrix converted to ASCII codes for conversion: ", to_array)
```

```python
print("Encrypted Data: ",encrypt)
```
